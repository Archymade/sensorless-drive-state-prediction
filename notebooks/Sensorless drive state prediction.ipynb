<h4> We will be fitting the data to an instance of the ExtraTree classifier.

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
tree = ExtraTreesClassifier(max_depth = 3, random_state = 42)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:, : -1]
y = data.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify = y)
```

```python
tree.fit(X_train, y_train)
yhat = tree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
print(classification_report(y_test, yhat))
```

<h4>From the results obtained above, it is seen that the model performs quite well, even without any serious hyperparametric tuning. Let us attempt to improve it through feature selection.

```python
feature_ranks = tree.feature_importances_
```

```python
feature_ranks
```

```python
feature_ranks.shape
```

```python
plt.figure(figsize = (20, 8))
sns.barplot(x = data.iloc[:, : -1].columns, y = 10**6*feature_ranks, color = 'blue')
```

```python
importances = pd.DataFrame({'Importance' : 1000*feature_ranks,
                            'Feature' : ['Feature {}'.format(i + 1) for i in data.iloc[:, : -1].columns]
                           })
```

```python
importances
```

```python
# Selecting the indexes of features which have importances greater than 1
i = 0
indexes = []

while i < 48:
    if list(1000*feature_ranks)[i] > 10:
        indexes.append(i)
    i += 1
```

```python
indexes
```

```python
### Index out selected features
X_train_new = X_train.iloc[:, indexes]
X_test_new = X_test.iloc[:, indexes]
```

```python
new_tree = ExtraTreesClassifier(max_depth = 3, random_state = 42)
```

```python
### Train model and test preformance
new_tree.fit(X_train_new, y_train)
new_yhat = new_tree.predict(X_test_new)
```

```python
print(classification_report(y_test, new_yhat))
```

<h4> We seem to have cut down on the not-so-important features. The ExtraTreesClassifier has an even more competent degree of accuracy. Perhaps this can be improved on by utilizing it as the basis for a boosting machine.<h4>
       
The selected boosting machine is an AdaBoostClassifier.

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
Ada_model = AdaBoostClassifier(base_estimator = ExtraTreesClassifier(max_depth = 3, random_state = 42), random_state = 42)
```

```python
Ada_model.fit(X_train_new, y_train)
yhat = Ada_model.predict(X_test_new)
print(classification_report(y_test, yhat))
```
